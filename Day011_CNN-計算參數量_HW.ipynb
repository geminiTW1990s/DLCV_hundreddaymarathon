```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import Input, Dense
from keras.models import Model
```

```python
##輸入照片尺寸==28*28*1
##都用一層，288個神經元

##建造一個一層的CNN層
classifier=Sequential()

##Kernel size 3*3，用32張，輸入大小28*28*1
#'''32張Kernel，大小為3*3，輸入尺寸為28*28*1'''
classifier.add(Convolution2D(filters=32, kernel_size=(3,3), input_shape=(28,28,1)))
##看看model結構
print(classifier.summary())
#'''理解輸出Total params為何==320'''

##建造一個一層的FC層
classifier=Sequential()
##輸入為28*28*1攤平==784
#'''輸入尺寸為28*28*1'''
#當只要一個element時但還是想要create tuple時就需要括號內加逗號, 第二個element可以空白沒關係, https://stackoverflow.com/questions/12876177/how-to-create-a-tuple-with-only-one-element
inputs = Input(shape=(784,))
##CNN中用了(3*3*1)*32個神經元，我們這邊也用相同神經元數量
#'''使用288個神經元'''
x=Dense(288)(inputs)
model = Model(inputs=inputs, outputs=x)
##看看model結構
print(model.summary())
#'''理解輸出Total params為何==226080'''

##大家可以自己變換設定看看參數變化
```
