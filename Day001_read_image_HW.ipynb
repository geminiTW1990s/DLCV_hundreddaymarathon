```python
img_path = './data/lena.png'
img = cv2.imread(img_path)
cv2_imshow(img)
```

```python
img = cv2.imread(img_path)
cv2_imshow(img[:,:,0])
```

```python
img = cv2.imread(img_path)
cv2_imshow(img[:,:,1])
```

```python
img = cv2.imread(img_path)
cv2_imshow(img[:,:,2])
```

```python
img = cv2.imread(img_path)
img_b = img
img_b[:,:,1] = np.zeros((512,512))
img_b[:,:,2] = np.zeros((512,512))
cv2_imshow(img_b)
```

```python
img = cv2.imread(img_path)
img_g = img
img_g[:,:,0] = np.zeros((512,512))
img_g[:,:,2] = np.zeros((512,512))
cv2_imshow(img_g)
```

```python
img = cv2.imread(img_path)
img_r = img
img_r[:,:,0] = np.zeros((512,512))
img_r[:,:,1] = np.zeros((512,512))
cv2_imshow(img_r)
```
